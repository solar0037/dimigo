```python
import pandas as pd
df = pd.DataFrame(
 {"temperature": [5, 8, 9, 15, 15, 18, 20],
 "precipitation": [0, 0, 5, 5, 8, 9, 15],
 "hiking": [0, 0, 0, 1, 1, 1, 0]}
)
df
```

```python
X_data = df[['temperature', 'precipitation']]
X_data
```

```python
y_data = df['hiking']
y_data
```

```python
X_data.shape, y_data.shape  # 모양 출력하기
```

```python
from sklearn.tree import DecisionTreeClassifier
# DecisionTreeClassifier 객체 생성
dt = DecisionTreeClassifier(random_state=11)
```

```python
# 모델 학습하기
dt.fit(X_data.values, y_data.values)
```

```python
# 결정 트리 그리기
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(7, 7))
plot_tree(dt,
 class_names=['stay', 'go'], 
 feature_names=df.columns,
 filled=True)
plt.show()
```

```python
# 특성 중요도
dt.feature_importances_
```

```python
# 오늘 기온과 강수량으로 등산을 갈지 예측하기
today_data = [[15, 5]]
pred = dt.predict(today_data)
pred
```
