```python
import pandas as pd

df = pd.DataFrame(
    {"income": [100, 200, 300, 400, 500],
     "card": [30, 57, 76, 140, 197]}
)
df
```

```python
X_data = df[['income']]
X_data
```

```python
y_data = df['card']
y_data
```

```python
# 산점도 그리기
import matplotlib.pyplot as plt

plt.xlabel('income')
plt.ylabel('card')
plt.scatter(df.income, df.card)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

# 모델 학습하기
lr.fit(X_data.values, y_data.values)
```

```python
# 회귀 계수 확인하기
lr.coef_, lr.intercept_
```

```python
# 선형회귀 예측하기
pred = lr.predict([[250]])
pred
```
