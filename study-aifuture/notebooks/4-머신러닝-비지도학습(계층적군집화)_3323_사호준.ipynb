```python
import pandas as pd
from sklearn import datasets
```

```python
# 아이리스 데이터 셋 가져오기
iris = datasets.load_iris()
iris
```

```python
# 데이터 셋 중에서 특징벡터만 가져오기
data = iris['data']
data[:5]  # 상위 5개 데이터 보기
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# 계층적 군집화 실행
# method: single, complete, average
matrix = linkage(y=data, method='complete', metric='euclidean')
matrix
```

```python
# 덴드로그램 그리기
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 10))
dendrogram(matrix, leaf_font_size=10)  # leaf_font_size: 글자크기
plt.show()
```

```python
# 클러스터 자르기
from scipy.cluster.hierarchy import fcluster

clusters = fcluster(matrix, t=4, criterion='distance')  # t: 덴드로그램 자르는 임계값
clusters
```

```python
# 실제 붓꽃 타겟값
iris.target
```

```python
# 군집화 결과 산점도 그리기
plt.xlabel(iris['feature_names'][0]) # sepal length
plt.ylabel(iris['feature_names'][2]) # petal length

cluster1 = data[clusters == 1]  # 첫번째 군집
plt.scatter(cluster1[:, 0], cluster1[:, 2], marker='o')

cluster2 = data[clusters == 2]  # 두번째 군집
plt.scatter(cluster2[:, 0], cluster2[:, 2], marker='^')

cluster3 = data[clusters == 3]  # 세번째 군집
plt.scatter(cluster3[:, 0], cluster3[:, 2], marker='*')

plt.show()
```

```python
# 원본 데이터 산점도 그리기
plt.xlabel(iris['feature_names'][0]) # sepal length
plt.ylabel(iris['feature_names'][2]) # petal length

plt.scatter(data[:, 0], data[:, 2], c=iris.target)

plt.show()
```
