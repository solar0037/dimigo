```python
# AND 게이트 구현
def AND(x1, x2):
  w1, w2 = 0.6, 0.6  # 가중치 w1, w2
  b = -0.85          # 진리표대로 나오도록 bias값을 적절히 주기
  result = x1*w1 + x2*w2 + b
  return 0 if result <= 0 else 1
```

```python
input_list = [(0, 0), (0, 1), (1, 0), (1, 1)]
for x1, x2 in input_list:
  result = AND(x1, x2)
  print(f'입력값: {x1} {x2} / AND 결과: {result}')
```

```python
# OR 게이트 구현
def OR(x1, x2):
  w1, w2 = 0.6, 0.6  # 가중치 w1, w2
  b = -0.25          # 진리표대로 나오도록 bias값을 적절히 주기
  result = x1*w1 + x2*w2 + b
  return 0 if result <= 0 else 1
```

```python
for x1, x2 in input_list:
  result = OR(x1, x2)
  print(f'입력값: {x1} {x2} / OR 결과: {result}')
```

```python
# NAND 게이트 구현
def NAND(x1, x2):
  return 1 if not AND(x1, x2) else 0
```

```python
for x1, x2 in input_list:
  result = NAND(x1, x2)
  print(f'입력값: {x1} {x2} / NAND 결과: {result}')
```

```python
# XOR 게이트 구현
def XOR(x1, x2):
  return AND(OR(x1, x2), NAND(x1, x2))
```

```python
for x1, x2 in input_list:
  result = XOR(x1, x2)
  print(f'입력값: {x1} {x2} / XOR 결과: {result}')
```

```python
import numpy as np
def init_network():
  network = {}
  network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
  network['b1'] = np.array([0.1, 0.2, 0.3])
  network['W2'] = np.array([[0.1, 0.2, 0.3]])
  network['b2'] = np.array([0.1])
  return network
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
def forward(network, x):
  W1, W2 = network['W1'], network['W2']
  b1, b2 = network['b1'], network['b2']
  a1 = np.dot(x, W1) + b1
  z1 = sigmoid(a1)          # (1, 2) * (2, 3) = (1, 3)
  a2 = np.dot(z1, W2.T) + b2  # (1, 3) * (3, 1) = (1, 1)
  y = sigmoid(a2)
  return y
```

```python
network = init_network()
x = np.array([1.0, 0.5])  # 입력값
y = forward(network, x)
print(y)
```
