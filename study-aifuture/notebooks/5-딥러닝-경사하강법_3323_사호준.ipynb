```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 공부시간 X와 성적 Y 데이터
# data = [[1, 10], [3, 30], [5, 50], [7, 70]]
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = np.array([i[0] for i in data])
y = np.array([i[1] for i in data])
x, y
```

```python
# 산점도 그리기
plt.scatter(x, y)
plt.show()
```

```python
lr = 0.01  # 학습률
W, b = 1, 0  # 가중치와 절편 초기화
# epochs = 100
epochs = 1000
```

```python
# 경사 하강법 구현
for i in range(epochs):
  y_hat = W * x + b  # 예측값
  error = y - y_hat  # 오차

  # 손실함수를 W, b로 미분한 값 (그래디언트)
  W_grad = -sum(x * error)
  b_grad = -sum(error)

  W = W - lr * W_grad  # 학습률을 곱해 W값 업데이트
  b = b - lr * b_grad  # 학습률을 곱해 b값 업데이트

  if i % 10 == 0:
    print(f'epoch={i}, 기울기={W}, 절편={b}')
```

```python
# 회귀선 추가하기
plt.scatter(x, y)
# plt.plot([1, 7], [1 * W + b, 7 * W + b])
plt.plot([1, 8], [1 * W + b, 8 * W + b])
plt.show()
```

```python
# 회귀식 직접 구하기
mean_x = np.mean(x)
mean_y = np.mean(y)

# 기울기 찾기
bottom = np.sum((x - mean_x) ** 2)
top = np.sum((x - mean_x) * (y - mean_y))
W2 = top / bottom

# 절편 찾기
b2 = mean_y - W2 * mean_x

print(W2, b2)  # 경사하강법에서 나온 값과 확인하기
```
