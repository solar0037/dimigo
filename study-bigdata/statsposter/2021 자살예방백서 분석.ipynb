```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./data/OECD 회원국 연령대별 자살률 현황.xlsx')
df
```

```python
oecd_avg = (df.iloc[34]['10대'])
oecd_avg
```

```python
df.drop(34, inplace=True)
```

```python
df.info()
```

```python
df.sort_values(by='10대', inplace=True)
df['국가'] = df['국가'] + " ['" + df['년도'].apply(lambda x: str(x)[2:]) + ']'
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = 14, 8
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
clrs = [
    'yellow',
    *['grey']*(len(df)-5),
    'skyblue',
    'deepskyblue',
    'blue',
    'purple'
]

plt.title('OECD 주요 회원국 청소년(10~19세) 자살률 [단위: 인구 10만 명당 명]')
sns.barplot(x='국가', y='10대', data=df, palette=clrs)
plt.xticks(rotation=-90)
plt.axhline(y=oecd_avg)
plt.show()
```
